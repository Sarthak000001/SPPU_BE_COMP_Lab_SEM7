Problem Statement:
Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2

```python
current_x = 2 #algorithm starts at x = 2 
rate = 0.01 #learning rate 
precision = 0.000001 #tells us when to stop the algorithm
precision_step_size = 1
max_iters = 10000 #maximum number of iterations
iters = 0 #iteration counter 
df = lambda x : 2*(x+3) #gradient of our function
```

```python
def function(x):
    return (x+3)**2
```

```python
x_values = [current_x]
y_values = [function(current_x)]
while precision_step_size > precision and iters < max_iters: ##When Previous Step SIze will be less than Precision then we will reach the global minima
    previous_x = current_x #store curr. value in previous_x
    current_x = current_x - (rate * df(previous_x)) #Formuala
    precision_step_size = abs(current_x - previous_x) #change in x
    iters = iters + 1 #iteration count
    x_values.append(current_x)
    y_values.append(function(current_x))
    print("Iteration",iters,"\n X value is ",current_x) 

print("The local minimum occurs at ",current_x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_range = np.linspace(-6,6,100)
y_range = function(x_range)

plt.figure(figsize=(10,6))
plt.plot(x_range,y_range,label='y = (x + 3)^2',color='blue')
plt.scatter(x_values,y_values,label='Gradient Descent Path',color='red')
plt.scatter(x_values[-1],y_values[-1],label='Local Minima',color='purple')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minima of y = (x + 3)^2")
plt.show()
```
